```python
import networkx as nx
import matplotlib.pyplot as plt

class GraphVisualization:
   
    def __init__(self):
        self.visual = []
          
    def addEdge(self, a, b):
        temp = [a, b]
        self.visual.append(temp)
          
    def visualize(self):
        G = nx.DiGraph()
        G.add_edges_from(self.visual)
        nx.draw_networkx(G)
        plt.show()
  

g = GraphVisualization()

# make a graph by adding edges 

g.addEdge(1, 2)
g.addEdge(2, 3)
g.addEdge(1, 3)
g.addEdge(1, 4)
g.addEdge(4, 3)
g.addEdge(5, 2)
g.addEdge(2, 6)
g.addEdge(5, 6)
g.addEdge(6, 2)

# visualize the graph using networkx
g.visualize()
```
